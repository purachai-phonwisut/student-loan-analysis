* กยศ. ย่อมาจาก กองทุนเงินให้กู้ยืมเพื่อการศึกษา เป็นหน่วยงานของรัฐที่จัดตั้งขึ้นตามพระราชบัญญัติกองทุนเงินให้กู้ยืมเพื่อการศึกษา พ.ศ. 2541 อยู่ภายใต้การกำกับดูแลของกระทรวงการคลัง มีวัตถุประสงค์หลักเพื่อสนับสนุนและส่งเสริมการศึกษาแก่นักเรียนหรือนักศึกษาที่ขาดแคลนทุนทรัพย์ โดยให้เงินกู้ยืมเพื่อเป็นค่าเล่าเรียน ค่าใช้จ่ายที่เกี่ยวเนื่องกับการศึกษา และค่าครองชีพ
